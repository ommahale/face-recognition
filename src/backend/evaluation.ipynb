```python
import os
```

```python
print(len(os.listdir('positive')))
print(len(os.listdir('negative')))
print(len(os.listdir('anchor')))
```

```python
input_img = os.listdir('positive')[:250]
input_img = ['./positive/'+x for x in input_img]
val_image = os.listdir('negative')[:125]
val_image = ['./negative/'+x for x in val_image]
val_image.extend(os.listdir('anchor')[:125])
val_image[125:] = ['./anchor/'+x for x in val_image[125:]]
```

```python
import numpy as np

pred = [0]*125 +[1]*125
```

```python
import pandas as pd
data = {'input_image':input_img, "validation_image":val_image, "target":pred}
dataset = pd.DataFrame(data= data)
```

```python
dataset
```

```python
from sklearn.metrics import classification_report
```

```python
import keras
import tensorflow as tf
class L1Dist(keras.layers.Layer):
    def __init__(self, **kwargs):
        super(L1Dist, self).__init__(**kwargs)

    def call(self, input_embedding, validation_embedding):
        return tf.math.abs(input_embedding[0] - validation_embedding[0])
```

```python
from keras.models import load_model
models = {
    'siamese':load_model('siamesemodel.keras', custom_objects={'L1Dist':L1Dist}),
    'vgg': load_model('vgg_model.keras', custom_objects={'L1Dist':L1Dist}),
    'mbnv2': load_model('mbnv2_model.keras', custom_objects={'L1Dist':L1Dist}),
}
```

```python


def preprocess(file_path):
    # Read in image from file path 
    byte_img = tf.io.read_file(file_path)
    
    # Load in the image 
    img = tf.io.decode_jpeg(byte_img)
    
    # Preprocessing steps - resizing the image to be 100x100x3
    img = tf.image.resize(img, (100,100))
    # Scale image to be between 0 and 1 
    img = img / 255.0
    img = tf.reshape(img,[-1,100,100,3])
    # Return image
    return img
```

```python
x1 = dataset['input_image']
x2 = dataset['validation_image']
```

```python
x1 = x1.apply(preprocess)
x2 = x2.apply(preprocess)
```

```python
x1
```

```python
x2
```

```python
list(zip(x1,x2))[0]
```

```python
y_pred = [models['siamese'].predict([input_image, validation_image])[0][0] for input_image, validation_image in list(zip(x1,x2))]
y_pred = [0 if i<0.5 else 1 for i in y_pred]
print(classification_report(dataset['target'], y_pred))
```

```python
y_pred = [models['vgg'].predict([input_image, validation_image])[0][0] for input_image, validation_image in list(zip(x1,x2))]
y_pred = [0 if i<0.5 else 1 for i in y_pred]
print(classification_report(dataset['target'], y_pred))
```

```python
y_pred = [models['mbnv2'].predict([input_image, validation_image])[0][0] for input_image, validation_image in list(zip(x1,x2))]
y_pred = [0 if i<0.5 else 1 for i in y_pred]
print(classification_report(dataset['target'], y_pred))
```
